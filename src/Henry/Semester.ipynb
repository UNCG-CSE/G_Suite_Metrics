```python
from datetime import datetime
from dateutil.parser import parse

import pandas as pd
data_path = "C:\\Users\\Henry\\Documents\\405-DataScience\\G_Suite_Metrics-Gsuite_Henry\\data\\Somewhat_cleaned_data.csv"
data = pd.read_csv(data_path)
data_path_write = 'C:\\Users\\Henry\\Documents\\405-DataScience\\G_Suite_Metrics\\data\\Semester Data\\'
```

Time in this dataframe is stored as a string, I am converting it to a time object. 

```python

data['time'] = pd.to_datetime(data.time)
```

```python
data
```

Seperating the main dataframe into semester dataframes and writing it to a csv file

```python
start = '2015-8-10T00:00:00-04:00'
end = '2015-12-10T00:00:00-05:00'
start = pd.to_datetime(start)
end = pd.to_datetime(end)
mask = (data['time'] >= start) & (data['time'] <= end)
fall2015 = data.loc[mask]
fall2015 = fall2015.reset_index()
fall2015=fall2015.drop(columns=['index']) 

```

```python

fall2015
```

```python
fall2015.to_csv(data_path_write + 'Fall2015.csv', index = False)
```
