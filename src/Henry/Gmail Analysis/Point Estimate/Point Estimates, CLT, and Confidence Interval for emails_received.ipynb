Lets find the difference between actual mean vs sample mean

```python
er.mean() - sample_er.mean()
```

Our point estimate mean is off by ~2255, lets apply the central limit theorem.


```python
pe = []         # Make empty list to hold point estimates

for x in range(200):         # Generate 200 samples
    sample = np.random.choice(a= er['emails_received'], size=300)
    pe.append(sample.mean())
```

```python
er.mean() - np.array(pe).mean()
```

```python
pd.DataFrame(pe).plot(kind="density")   
```

Now the mean is much closer and the  distribution looks normal, now lets find the confidence levels.  (Please note I did confidence levels before in the estimators folder). I am calculating for 95% accuracy.

```python
er_size = er.size
er_mean = er.mean()
z_critical = stats.norm.ppf(q = 0.975)  # Get the z-critical value*
                        

er_std = er.std()  # Get the population standard deviation

margin_of_error = z_critical * (er_std/math.sqrt(er_size))

confidence_interval = (er_mean - margin_of_error,
                       er_mean + margin_of_error)  

print("Sample Mean:")
print(er_mean)

print("Confidence interval:")
print(confidence_interval)
```

```python
np.array(pe).mean()
```

We can see the sample mean is within the margin of error
