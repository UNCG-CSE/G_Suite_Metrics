```python
es_size = es.size
es_mean = es.mean()
z_critical = norm.ppf(q = 0.975)  # Get the z-critical value*
                        

es_std = es.std()  # Get the population standard deviation

margin_of_error = z_critical * (es_std/math.sqrt(es_size))

confidence_interval = (es_mean - margin_of_error,
                       es_mean + margin_of_error)  

print("Sample Mean:")
print(es_mean)

print("Confidence interval:")
print(confidence_interval)
```
