The means for active users in the month of december vs overall mean for active users appears to be different, lets try some hypothesis testing. The hypothesis test that would work best for this question will be the One-Sample T-Test. It tests whether the sample mean is different from the population mean. For this case, we are testing if there is a significance between the month of december (the sample mean) vs the overall mean in the dataset (population mean).
The null hypothesis (h0) - Active users in the month of December is similar
the alternative hypothesis(h1) - active users in the month of December is not similar. The p-value threshold is p = 0.05

```python
stats.ttest_1samp(a= dec,               # Sample data
                 popmean= odu.mean())  # Pop mean
```

The p-value is significanntly lower then the p-value threshold, thus we can reject the null hypothesis. Thus the mean values of the sample mean (dec) and the population mean (odu) are different. 

```python
odu_size = odu.size
odu_mean = odu.mean()
z_critical = norm.ppf(q = 0.975)  # Get the z-critical value*
                        

odu_std = odu.std()  # Get the population standard deviation

margin_of_error = z_critical * (odu_std/math.sqrt(odu_size))

confidence_interval = (odu_mean - margin_of_error,
                       odu_mean + margin_of_error)  

print("Sample Mean:")
print(odu_mean)

print("Confidence interval:")
print(confidence_interval)
```

```python
dec_size = dec.size
dec_mean = dec.mean()
z_critical = norm.ppf(q = 0.975)  # Get the z-critical value*
                        

dec_std = dec.std()  # Get the population standard deviation

margin_of_error = z_critical * (dec_std/math.sqrt(dec_size))

confidence_interval = (dec_mean - margin_of_error,
                       dec_mean + margin_of_error)  

print("Sample Mean:")
print(dec_mean)

print("Confidence interval:")
print(confidence_interval)
```
