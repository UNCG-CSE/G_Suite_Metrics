```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import datetime
df = pd.read_csv('Gcalender.csv')
```

```python

df['time'] =  pd.to_datetime(df['time'], format='%Y%m%d %H:%M:%S.%f', utc=True)
dailyDf = df[df['metric_name'].str.match('google.calendar:num_1day_active_users')]
newDaily = dailyDf.drop([3354])
```

```python
dailyDf.index = dailyDf['time'] 
dailyDf = dailyDf[dailyDf['metric_name'].str.match('google.calendar:num_1day_active_users')]
weekly=dailyDf.resample('W').sum()
weekly.head()
weekly.reset_index(inplace=True)
weekly.describe()
```

```python
weekly.hist()
```

```python
weekly['time']=pd.to_datetime(weekly['time'],utc=True)
weekly.set_index('time',inplace=True)
weekly.plot()
```

```python
monthly=dailyDf.resample('M').sum()

monthly.reset_index(inplace=True)

monthly.describe()
```

```python
monthly.hist()
```

```python
monthly['time']=pd.to_datetime(monthly['time'],utc=True)
monthly.set_index('time',inplace=True)
monthly.plot()
```

```python
yearly=dailyDf.resample('Y').sum()

yearly.reset_index(inplace=True)
yearly.describe()
```

```python
yearly.hist()
```

```python
yearly['time']=pd.to_datetime(yearly['time'],utc=True)
yearly.set_index('time',inplace=True)
yearly.plot()
```
