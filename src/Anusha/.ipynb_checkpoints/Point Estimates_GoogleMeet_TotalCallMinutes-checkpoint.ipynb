Central Limit Theorem - CLT
The distribution of many sample means, known as a sampling distribution, will be normally distributed.

To illustrate, let's create a sampling distribution by taking 300 samples from our call minutes data and then making 300 point estimates of the mean:

```python
np.random.seed(10)

point_estimates = []         # Make empty list to hold point estimates

for x in range(300):         # Generate 200 samples
    sample = np.random.choice(a= data2['TotalCallMinutes'], size=300)
    point_estimates.append( sample.mean() )
    
f, ax = plt.subplots(figsize=(12, 10))
sns.distplot( point_estimates, hist=False, rug=True); # Plot sample mean density
```

The sampling distribution appears to be roughly normal, despite the exponential distribution that the samples were drawn from. 
In addition, the mean of the sampling distribution approaches the true mean, there is only a difference of 3 minutes:

```python
data2.mean() - np.array(point_estimates).mean()
```

A point estimate can give you a rough idea of a distribution parameter like the mean, but estimates are prone to error and taking multiple samples to get improved estimates may not be feasible.
A confidence interval is a range of values above and below a point estimate that captures the true distribution parameter at some predetermined confidence level. 

Let's calculate a 95% confidence for our mean point estimate:

```python
np.random.seed(10)

sample_size = 400
sample = np.random.choice(a= data2['TotalCallMinutes'], size = sample_size)
sample_mean = sample.mean()

z_critical = stats.norm.ppf(q = 0.975)  # Get the z-critical value*

print("z-critical value:")              # Check the z-critical value
print(z_critical)                        

pop_stdev = data2.std()  # Get the population standard deviation

margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size))

confidence_interval = (sample_mean - margin_of_error,
                       sample_mean + margin_of_error)  

print("Sample Mean:")
print(sample_mean)

print("Confidence interval:")
print(confidence_interval)
```

Notice that the confidence interval we calculated is very close the true call minutes mean of 1148.051565

```python
np.random.seed(12)

sample_size = 400

intervals = []
sample_means = []

for sample in range(25):
    sample = np.random.choice(a= data2['TotalCallMinutes'], size = sample_size)
    sample_mean = sample.mean()
    sample_means.append(sample_mean)

    z_critical = stats.norm.ppf(q = 0.975)  # Get the z-critical value*         

    pop_stdev = data2.std()  # Get the population standard deviation

    stats.norm.ppf(q = 0.025)

    margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size))

    confidence_interval = (sample_mean - margin_of_error,
                           sample_mean + margin_of_error)  
    
    intervals.append(confidence_interval)
```

```python
plt.figure(figsize=(12,10))

plt.errorbar(x=np.arange(0.1, 25, 1), 
             y=sample_means, 
             yerr=[(top-bot)/2 for top,bot in intervals],
             fmt='o')

plt.hlines(xmin=0, xmax=25,
           y=1148.051565, 
           linewidth=2.0,
           color="red")
```

Notice that in the prior plot, 24 samples of the 95% confidence intervals overlap the red line marking the true mean. 
- This is to be expected: since a 95% confidence interval captures the true mean 95% of the time, we'd expect our interval to miss the true mean 5% of the time. (25 samples)

If you don't know the standard deviation of the population, you have to use the standard deviation of your sample as a stand in when creating confidence intervals.

Since the sample standard deviation may not match the population parameter the interval will have more error when you don't know the population standard deviation.
To account for this error, we use what's known as a t-critical value instead of the z-critical value.

The t-critical value is drawn from what's known as a t-distribution
Let's take a new, smaller sample and then create a confidence interval without the population standard deviation, using the t-distribution:

```python
np.random.seed(10)

sample_size = 25
sample = np.random.choice(a= data2['TotalCallMinutes'], size = sample_size)
sample_mean = sample.mean()

t_critical = stats.t.ppf(q = 0.975, df=24)  # Get the t-critical value*

print("t-critical value:")                  # Check the t-critical value
print(t_critical)                        

sample_stdev = sample.std()    # Get the sample standard deviation

sigma = sample_stdev/math.sqrt(sample_size)  # Standard deviation estimate
margin_of_error = t_critical * sigma

confidence_interval = (sample_mean - margin_of_error,
                       sample_mean + margin_of_error)  

print("Confidence interval:")
print(confidence_interval)
```

t-critical value is larger than the z-critical value

If you have a large sample, the t-critical value will approach the z-critical value so there is little difference between using the normal distribution vs. the t-distribution:

```python
# Check the difference between critical values with a sample size of 400
             
stats.t.ppf(q=0.975, df= 399) - stats.norm.ppf(0.975)  
```

```python
stats.t.interval(alpha = 0.95,              # Confidence level
                 df= 24,                    # Degrees of freedom
                 loc = sample_mean,         # Sample mean
                 scale = sigma)             # Standard deviation estimate
```
