```python
data_path2 = "C:/Users/Anusha/Documents/Copied on 9_26_19/Fall 2019/Data Science/Project/GoogleMeet.csv"
noofcalls = pd.read_csv(data_path2)
noofcalls['time'] = pd.to_datetime(noofcalls['time'],utc=True)
noofcalls.set_index('time', inplace=True)
noc = noofcalls['num_calls']
```

Kernel Density Estimation - KDE

```python
from scipy.stats import kde
noc.hist(normed=True, bins=20, grid=False, ec='black', figsize=(12,8))
x = np.linspace(0, noc.max())
density = kde.gaussian_kde(noc, bw_method=None)
xgrid = np.linspace(x.min(), x.max(), 100)

plt.plot(x, density(x), 'y--')
```
