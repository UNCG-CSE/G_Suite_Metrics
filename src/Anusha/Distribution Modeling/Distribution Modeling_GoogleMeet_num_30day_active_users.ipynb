Kernel Density Estimation - KDE

```python
from scipy.stats import kde
nac.hist(normed=True, bins=20, grid=False, ec='black', figsize=(12,8))
x = np.linspace(0, nac.max())
density = kde.gaussian_kde(nac, bw_method=None)
xgrid = np.linspace(x.min(), x.max(), 100)

plt.plot(x, density(x), 'y--')
```
