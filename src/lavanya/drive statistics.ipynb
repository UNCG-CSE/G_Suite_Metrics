```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('google_drive_modified.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
# no of nan values in a column
```

```python
# dropping 3 columns with all nan values
df.drop(['google.drive:num_owned_items_with_visibility_unknown_added','google.drive:num_owned_items_with_visibility_unknown_delta','google.drive:num_owned_items_with_visibility_unknown_removed'], axis =1)
```

Finding all the attributes whose values values are totally null

```python
df.isnull().sum()
```

Replacing the nan values with the mean

```python
df = df.fillna(df.mean())
```

```python
#df.columns = df.columns.str.rstrip('google.drive:num_')
df.rename(columns=lambda x: x.replace('google.drive:num_', ''), inplace=True)
df.head()
```

```python
df.mean()
#mean of each column
```

```python
df.median()
#median of each column
```

```python
df.to_csv( 'google_drive_nanreplaced.csv', index = False)
df.head()
```

Plotting the inter quartile range for the attribute 1day_active_users

```python
df.boxplot(column="1day_active_users",
               return_type='axes',
               figsize=(10,10))

plt.text(x=0.74, y=5700, s="3rd Quartile")
plt.text(x=0.8, y=3300, s="Median")
plt.text(x=0.75, y=2100, s="1st Quartile")
plt.text(x=0.9, y=100, s="Min")
plt.text(x=0.9, y=9100, s="Max")
plt.text(x=0.7, y=3500, s="IQR", rotation=90, size=25)
```

Plotting th einterquartile range for the attribute 30day active users

```python
df.boxplot(column="30day_active_users",
               return_type='axes',
               figsize=(10,10))

plt.text(x=0.74, y=18000, s="3rd Quartile")
plt.text(x=0.8, y=16900, s="Median")
plt.text(x=0.75, y=13500, s="1st Quartile")
plt.text(x=0.9, y=9000, s="Min")
plt.text(x=0.9, y=20000, s="Max")
plt.text(x=0.7, y=16000, s="IQR", rotation=90, size=25)
```
